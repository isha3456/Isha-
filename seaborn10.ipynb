```python
#11.Students Performance Dashboard

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "students":["Amit","Sneha","Priya","Anju","Gouri"],
    "marks":[70,85,93,90,69]
}

df=pd.DataFrame(data)
df.to_csv("students2.csv",index=False)

sns.barplot(x="students",y="marks",data=df)
plt.title("Students Performance Dashboard")
plt.show()

print("Highest Scoring:",df.loc[df["marks"].idxmax(),"students"])
```

```python
#13. Employee Productivity Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "employee":["Amit","Sneha","Priya","Anju","Gouri"],
    "tasks":[25,40,35,32,28]
}

df=pd.DataFrame(data)
df.to_csv("task.csv",index=False)

sns.barplot(x="employee",y="tasks",data=df)
plt.title("Employee Productivity Analysis")
plt.show()

print("Most Productive Employee:",df.loc[df["tasks"].idxmax(),"employee"])
```

```python
#18. Weather Distribution

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "Tempersture":[28,30,31,29,35,33,27,32,34,30]
}

df=pd.DataFrame(data)
df.to_csv("temperature.csv",index=False)

sns.histplot(df["Tempersture"],kde=True)
plt.title("Weather Distribution")
plt.show()
```

```python
#19. Hospital Patient Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "Patients":["P1","P2","P3","P4","P5"],
    "WaitingTime":[15,25,10,30,20]
}

df=pd.DataFrame(data)
df.to_csv("patients.csv",index=False)

sns.barplot(x="Patients",y="WaitingTime",data=df)
plt.title("Hospital Patient Analysis")
plt.show()

print("Average Waiting Time:",df["WaitingTime"].mean())
```

```python
#20. Social Media Engagement Report

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data={
    "Post":["Post1","Post2","Post3","Post4","Post5"],
    "Likes":[200,350,250,220,400],
    "Comments":[25,30,20,35,22]
}

df=pd.DataFrame(data)
df.to_csv("engagement.csv",index=False)

sns.barplot(x="Post",y="Likes",data=df)
plt.title("Social Media Likes")
plt.show()

sns.barplot(x="Post",y="Comments",data=df)
plt.title("Social Media Comments")
plt.show()
```
