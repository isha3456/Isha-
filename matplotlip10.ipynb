```python
#1. Student Marks Visualization

import matplotlib.pyplot as plt

students=["Isha","Sayali","Priya","Prachi","Ashu"]
marks=[90,92,78,84,68]

plt.bar(students,marks)
plt.title("Student Marks Visualization")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr","May","Jun"]
sales=[4000,6000,7000,8000,5000,10000]

plt.plot(months,sales,marker="o")
plt.title("Sales Analysis")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

employees=["Ram","Sita","Lucky","Riya","Samarth"]
salary=[50000,30000,60000,40000,70000]

plt.bar(employees,salary)
plt.title("Employee Salary Analysis")
plt.xlabel("Salary")
plt.ylabel("Employees")
plt.show()
```

```python
import matplotlib.pyplot as plt

students=["Ram","Sita","Lucky","Riya","Samarth"]
attendance=[90,92,60,70,78]

plt.bar(students,attendance)
plt.title("Student Attendance Analysis")
plt.xlabel("Students")
plt.ylabel("Attendance(%)")
plt.show()

print("Students Below 75% Attendance:")
for i in range(len(attendance)):
  if attendance[i]<75:
    print(students[i])
```

```python
import matplotlib.pyplot as plt

product=["Laptop","Headphones","Keyboard","Mouse","Mobile"]
sales=[120,150,200,80,100]

plt.bar(product,sales)
plt.title("Product Sales Analysis")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

top_product=product[sales.index(max(sales))]
print("Top Product:",top_product)
```

```python
import matplotlib.pyplot as plt

days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
steps=[4000,5000,6000,2000,1000,7000,8000]

plt.plot(days,steps,marker="o")
plt.title("Fitness Activity Tracker")
plt.xlabel("Days")
plt.ylabel("Steps")
plt.show()

active_day=days[steps.index(max(steps))]
print("Most Active Day:",active_day)
```

```python
import matplotlib.pyplot as plt

books=["Python","Artificial Intelligence","Machine Learning","Java Programming","Data Science"]
issued=[60,45,50,55,70]

plt.bar(books,issued)
plt.title(" Library Book Analysis")
plt.xlabel("Books")
plt.ylabel("Issued")
plt.show()

popular_book=books[issued.index(max(issued))]
print("Most Popular Book:",popular_book)

```
