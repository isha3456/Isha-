```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

all_patients_data = [] # Initialize as a list to store multiple patient dictionaries
n= int(input("Enter the number of patients: "))

for i in range(n):
    print(f"\n Enter details of patient{i+1}:")

    pid= int(input("Enter the patient ID: "))
    name= input("Enter the patient name: ")
    age= int(input("Enter the patient age: "))
    temperature= float(input(" Body temperature: "))

    if temperature>=100:
        status="Admit"
    else:
        status="Discharge"

    patient_data = { # Create a dictionary for the current patient
        "Patient_ID": pid,
        "Name": name,
        "Age": age,
        "Temperature": temperature,
        "Status": status
    }
    all_patients_data.append(patient_data) # Append the dictionary to the list

# After the loop, create the DataFrame from the list of patient dictionaries
df= pd.DataFrame(all_patients_data)

print("\n-------------Patient Records------------- ")
print(df)

temp_array=np.array(df["Temperature"])

print("\n-------------Temperature Records------------- ")
print("Highest Temperature:",np.max(temp_array))
print("Lowest Temperature:",np.min(temp_array))
print("Average Temperature:",np.mean(temp_array))

admit_count=len(df[df["Status"]=="Admit"])
discharge_count=len(df[df["Status"]=="Discharge"])

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Temperature"])
plt.title("Hospital Patient Analytics Dashboard")
plt.xlabel("Patient Name")
plt.ylabel("Temperature") # Corrected from plt.ylable
plt.grid(True)
plt.show()
```
