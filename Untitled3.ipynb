```python
import numpy as np
import pandas as pd

students=["Isha","Sanika","Omkar","Aditi","Jay"]
marks=np.array([90,80,70,68,75])
grades=[]

for mark in marks:
  if mark>=90:
    grades.append("A")
  elif mark>=75:
    grades.append("B")
  elif mark>=50:
    grades.append("C")
  else:
    grades.append("Fail")

df=pd.DataFrame({
    "Student": students,
    "Marks": marks,
    "Grades": grades
})

average_marks=np.mean(marks)
highest_marks=np.max(marks)

print("\n Student Performance Report")
print(df)

print("\n Average Marks",average_marks)
print("Highest Marks",highest_marks)
```

```python
import numpy as np
import pandas as pd

houses=["House1","House2","House3","House4","House5"]
units=np.array([1000,1500,1200,1800,2000])
bills=[]

for unit in units:
  if unit>=1500:
    bill=unit * 7
  elif unit>=1000:
    bill=unit * 5
  else:
    bill=unit * 10
  bills.append(bill)

df=pd.DataFrame({
     "House": houses,
     "Units": units,
     "Bills": bills
     })
print(df)
print("\n Average Usage:",np.mean(units))
```

```python
import numpy as np
import pandas as pd
students=["Isha","Sanika","Omkar","Aditi","Jay"]
attendance=np.array([90,60,58,75,85])
status=[]

for attend in attendance:
  if attend>=90:
    status.append("Excellent")
  elif attend>=75:
    status.append("Eligible")
  else:
    status.append("Not Eligible")

df=pd.DataFrame({
    "Student": students,
    "Attendance": attendance,
    "Status": status
})

print(df)
print("Average Attendance:",np.mean(attendance))
```

```python
import numpy as np
import pandas as pd

products=["Laptop","Mouse","Keyboard","Printer","Antivirus"]
quantity=np.array([50,10,5,30,2])
status=[]
for qty in quantity:
  if qty>=20:
    status.append("Sufficient")
  elif qty>=5:
    status.append("Running Low")
  else:
    status.append("Records needed")

df=pd.DataFrame({
    "Product": products,
    "Quantity": quantity,
    "Status": status
})

print(df)

```

```python
import numpy as np
import pandas as pd

days=["Mon","Tue","Wed","Thu","Fri","sat"]
steps=np.array([12000,14000,8000,6000,15000,9000])
performance=[]

for step in steps:
  if step>=10000:
    performance.append("Excellent")
  elif step>=7000:
    performance.append("Good")
  else:
    performance.append("Need Improvement")

df=pd.DataFrame({
    "Day": days,
    "Steps": steps,
    "Performance": performance
})

print(df)
print("\n Total Steps:",np.sum(steps))
print("Average Steps:",np.mean(steps))
```

```python
import pandas as pd

df=pd.read_csv("patients.csv")
status=[]

for bp in df["BP"]:
  if bp<130:
    status.append("Normal")
  elif bp<150:
    status.append("Moderate")
  else:
    status.append("High")

df["Health Status"]=status
print(df)
```

```python
import pandas as pd

df=pd.read_csv("tanks.csv")
status=[]

for level in df["WaterLevel"]:
  if level>=80:
    status.append("Full")
  elif level>=40:
    status.append("Medium")
  else:
    status.append("Low")

df["Status"]=status

print(df)
```

```python
import pandas as pd

df=pd.read_csv("delivery.csv")

df.columns = df.columns.str.strip().str.lower()
status=[]

for time in df["time"]: # Changed to df["time"] to correctly access the lowercase column name
  if time<=25:
    status.append("Fast")
  elif time<=45:
    status.append("On Time")
  else:
    status.append("Delayed")

df["Performance"]=status

print(df)
```
